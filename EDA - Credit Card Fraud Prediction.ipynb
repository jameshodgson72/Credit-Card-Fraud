## Fraud and categorical variables

Here I've created a couple of functions to help with drawing the plots I want below.  The first one extracts the data

```python
def query_how_fraud_varies_by_cat_variable( variable ):
    """Function to query my EDA database and return a data frame of the results"""
    # build the query string using the passed in argument "variable"
    query_string=\
    f'select t.{variable}, 100*(count(*)/var_cnt) as non_fraud, 100*(1-(count(*)/var_cnt)) as fraud, var_cnt as samples '\
    f'from ( '\
    f'select {variable}, cast(count(*) as float) as var_cnt '\
    f'from trans ' \
    f'group by {variable} '\
    f') c, trans t '\
    f'where c.{variable} = t.{variable} '\
    f'and t.tx_is_fraud = 0 '\
    f'group by t.{variable} '\
    f'order by t.{variable}'    
    
    return query(query_string)
```

```python
def plot_how_fraud_varies_by_cat_variable( variable, query_data ):
    """ Function used to plot the results when using the above query on my transaction database 
    and look at how the proportion of fraud varies across different categories of any specified 
    categorical variable passed in as an argument to this function."""
    
    # find the minimum value of fraud percentage (+5%) to limit the y axis
    min_fraud = 100.0 - (1.05*query_data.fraud.max())
    max_samples = 1.05*query_data.samples.max()
    
    # plot the query data.  Dynamically set the axis labels, title, etc.
    figure, ax1 = plt.subplots(figsize=(15, 6))
    ax2=ax1.twinx()

    query_data.plot(kind='bar', stacked=True, x=variable, y=['non_fraud','fraud'], alpha=0.85, ax=ax1 )
    sns.lineplot(data=query_data, x=ax1.get_xticks(), y='samples', ax=ax2, color='r', lw=1)

    ax1.set_xticklabels(query_data[variable].values, rotation=90)

    ax1.legend(loc='upper left')
    ax2.legend(['samples'], loc='upper right')
    ax1.set_ylim(min_fraud,100)
    ax2.set_ylim(0,max_samples)
    
    ax1.set_xlabel(data_dictionary[variable])
    ax1.set_ylabel("percentage")
    plt.title(f'How fraud varies with {data_dictionary[variable]}')
    plt.show()
```

## Fraud and continuous variables

Functions below are used to create histograms of continuous variables, splitting the samples in each histogram bin according to how many are fraud and how many are not.  Number of samples in each bin is also included in the extract and plots.

```python
def query_how_fraud_varies_by_cont_variable( variable ):
    """ function to create a histogram with continuous variables, and then to calculate the average of the
    target variable within each histogram bin."""
    num_bins=20 # number of histogram bins (at most - there may be gaps where there are no samples)

    query_string=\
    f'select '\
    f'case '\
    f'when {variable} >= 0 then (cast({variable}/((max_amt-min_amt)/{num_bins}.0) as int))*((max_amt-min_amt)/{num_bins}.0) '\
    f'when {variable} < 0 then (cast({variable}/((max_amt-min_amt)/{num_bins}.0) as int)-1)*((max_amt-min_amt)/{num_bins}.0) '\
    f'end as bin_min, '\
    f'case '\
    f'when {variable} >= 0 then (cast({variable}/((max_amt-min_amt)/{num_bins}.0) as int)+1)*((max_amt-min_amt)/{num_bins}.0) '\
    f'when {variable} < 0 then (cast({variable}/((max_amt-min_amt)/{num_bins}.0) as int))*((max_amt-min_amt)/{num_bins}.0) '\
    f'end as bin_max, '\
    f'100*avg(tx_is_fraud) as fraud, 100*(1-avg(tx_is_fraud)) as non_fraud, count(*) as samples '\
    f'from '\
    f'(select max({variable}) as max_amt from trans ) max, '\
    f'(select min({variable}) as min_amt from trans ) min, '\
    f' trans '\
    f'group by bin_max'

    return query(query_string)
```

```python
def plot_how_fraud_varies_by_cont_variable( variable, query_data ):
    """ function to create a plot using the supplied variable and query data (data as produced by the above function)."""
    
    # create a new column which is the mid-point of each bin in the data
    query_data['bin_mid'] = round((query_data.bin_max + query_data.bin_min)/2)
    
    # work out the minimum percentage to display on the primary y axis
    min_fraud=(100.0 - 1.05*query_data.fraud.max())#query_data.non_fraud.min()
    max_samples=1.05*query_data.samples.max()
    
    fig,ax1=plt.subplots(figsize=(15,6))
    ax2=ax1.twinx()
    query_data.plot(kind='bar', stacked=True, y=['non_fraud','fraud'], x='bin_mid', ax=ax1, alpha=0.85 )
    ax1.set_ylabel('percentage')
    ax1.legend(loc='upper left')
    sns.lineplot(data=query_data, x=ax1.get_xticks(), y='samples', ax=ax2, color='r', lw=2)
    ax2.legend(['samples'], loc='upper right')
    ax2.set_ylabel('Number of samples')
    ax1.set_ylim(min_fraud,100)
    ax2.set_ylim(0,max_samples)

    ax1.set_xlabel(data_dictionary[variable])
    plt.title(f'How fraud varies with {data_dictionary[variable]}')
    plt.show()
```

```python
# Alternative to the above functions using just Pandas instead of SQL.  Small advantage of this version
# is that it will create a bin for every interval, even when there are no samples in the interval.
def draw_hist_for_cont_variable( variable ):
    num_bins=20
    bin_mids=[]
    pc_fraud=[]
    pc_non_fraud=[]
    samples=[]
    
    range_min = X[variable].min()
    range_max = X[variable].max()
    bin_size = (range_max - range_min)/num_bins
    
#    print(f'min {range_min}, max {range_max}, size {bin_size}')
    
    for bin_num in range(0,num_bins):
        bin_min = range_min + (bin_num * bin_size)
        bin_max = range_min + ((bin_num+1) * bin_size)
        bin_mids.append( round(((bin_min + bin_max)/2),1) )
        
#        print( f'On bin {bin_num}, bin_min={bin_min}, bin_max={bin_max}')
        if( bin_num+1 == num_bins ):
            # use <= to compare against bin_max on the last bin so as not to miss any samples
            # at the top of the range.
            X_sub = X[(X[variable] >= bin_min) & (X[variable] <= bin_max)]
        else:
            X_sub = X[(X[variable] >= bin_min) & (X[variable] < bin_max)]
        
        bin_samples = X_sub.shape[0]
        
        if bin_samples != 0:
#            print(X_sub['tx_is_fraud'].value_counts())
            fraud_samples = X_sub[X_sub['tx_is_fraud'] == 1].user_id.count()
            non_fraud_samples = X_sub[X_sub['tx_is_fraud'] == 0].user_id.count()
            pc_fraud.append(100*fraud_samples/bin_samples)
            pc_non_fraud.append(100*non_fraud_samples/bin_samples)        
            samples.append(bin_samples)  
        else:
            pc_fraud.append(0)
            pc_non_fraud.append(0)            
            samples.append(bin_samples)            

    query_data = pd.DataFrame(data={'bin_mid':bin_mids, 'fraud':pc_fraud, 'non_fraud':pc_non_fraud, 'samples':samples})
    
    # work out the minimum percentage to display on the primary y axis
    min_fraud=(100.0 - 1.05*query_data.fraud.max())
    max_samples=1.05*query_data.samples.max()
    
    fig,ax1=plt.subplots(figsize=(15,6))
    ax2=ax1.twinx()
    query_data.plot(kind='bar', stacked=True, y=['non_fraud','fraud'], x='bin_mid', ax=ax1, alpha=0.85 )
    ax1.set_ylabel('percentage')
    ax1.legend(loc='upper left')
    sns.lineplot(data=query_data, x=ax1.get_xticks(), y='samples', ax=ax2, color='r', lw=2)
    ax2.legend(['samples'], loc='upper right')
    ax2.set_ylabel('Number of samples')
    ax1.set_ylim(min_fraud,100)
    ax2.set_ylim(0,max_samples)

    ax1.set_xlabel(data_dictionary[variable])
    plt.title(f'How fraud varies with {data_dictionary[variable]}')
    plt.show()    
  
# draw_hist_for_cont_variable('cc_credit_limit')
```
